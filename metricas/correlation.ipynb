```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'new_features\fourier_features.csv')
df = df[df['Fault type'] == 1]
# df = df[df['Fault resistance'] == 10]
# df = df[df['Pen'] == 75]
# df = df[df['Incidence angle'] == 90.1]

print(df['Fault resistance'].unique())

x = df['Fault distance'].values
y = df['Pos-Va -  Energy'].values
x, y = zip(*sorted(zip(x, y)))

# plt.figure(figsize=(16, 5))
# plt.plot(x,y, '-o')


correlation_with_fault_location = df.corr(method='pearson')['Fault distance'].drop('Fault distance').drop('Unnamed: 0').drop('Fault location')
correlation_with_fault_location = correlation_with_fault_location.sort_values()

threshold = 0.5
correlation_with_fault_location = correlation_with_fault_location[(correlation_with_fault_location > threshold)|(correlation_with_fault_location < -threshold)]

plt.figure(figsize=(10, 6))
sns.barplot(x=correlation_with_fault_location.index, y=correlation_with_fault_location.values, palette='coolwarm')
plt.xticks(rotation=90)
plt.title('Fourier - Pearson Correlation of Features with Fault location')
plt.xlabel('Correlation Coefficient')
plt.ylabel('Features')
plt.grid(True)
plt.show()
```
